````
import glob
import os
import matplotlib.pyplot as plt
#!pip install q keras==2.3.1

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint
##
from tensorflow.keras.layers import BatchNormalization

#from keras.layers.normalization import BatchNormalization
from tensorflow.keras.regularizers import l2
````

````
training_path=r'/gdrive/MyDrive/Colab Notebooks/horse-or-human/train'
validation_path=r'/gdrive/MyDrive/Colab Notebooks/horse-or-human/validation'
````

**Data Generator**

````
training_datagen = ImageDataGenerator(rescale=1./255,
                                      rotation_range=40,
                                      width_shift_range=0.2,
                                      height_shift_range=0.2,
                                      shear_range=0.2,
                                      zoom_range=0.2,
                                      horizontal_flip=True,
                                      fill_mode='nearest')

training_data =  training_datagen.flow_from_directory(training_path, 
                                      target_size=(150, 150), 
                                      batch_size=32,
                                      class_mode='binary')

validation_datagen=ImageDataGenerator(rescale=1/.255,
                                      rotation_range=40,
                                      width_shift_range=0.2,
                                      height_shift_range=0.2,
                                      shear_range=0.2,
                                      zoom_range=0.2,
                                      horizontal_flip=True,
                                      fill_mode='nearest')

validation_data=validation_datagen.flow_from_directory(validation_path,
                                                       target_size=(150,150),
                                                       batch_size=32,
                                                       class_mode='binary')
````

**Plotting The Images**

````
def plotImage(image_arr):
  fig,axes=plt.subplots(1,5,figsize=(20,20))
  axes=axes.flatten()
  for img,ax in zip(image_arr,axes):
    ax.imshow(img)
  plt.tight_layout()
  plt.show()

images=[training_data[0][0][0] for i in range(5)]
plotImage(images)
````

````
model=tf.keras.models.Sequential([
                               #keras.layers.Conv2D(filters=32, kernel_size=7, input_shape=[150,150,3],kernel_regularizer=12(l=0.01)),
                               tf.keras.layers.Conv2D(filters=32, kernel_size=7, input_shape=[150,150,3],kernel_regularizer=l2(l=0.01)),
                               BatchNormalization(),
                               tf.keras.layers.MaxPooling2D(pool_size=(2,2)),

                               tf.keras.layers.Conv2D(filters=64,kernel_size=5),
                               BatchNormalization(),
                               tf.keras.layers.MaxPooling2D(pool_size=(2,2)),

                               
                               

                               tf.keras.layers.Conv2D(filters=64,kernel_size=5),
                               BatchNormalization(),
                               tf.keras.layers.MaxPooling2D(pool_size=(2,2)),

                               tf.keras.layers.Conv2D(filters=128,kernel_size=3),
                               BatchNormalization(),
                               tf.keras.layers.MaxPooling2D(pool_size=(2,2)),

                               tf.keras.layers.Conv2D(filters=256, kernel_size=3),
                               BatchNormalization(),
                               tf.keras.layers.MaxPooling2D(pool_size=(2,2)),


                               tf.keras.layers.Flatten(),
                               
                               tf.keras.layers.Dense(units=128,activation='relu'),
                               BatchNormalization(),

                               tf.keras.layers.Dropout(0.5),
                               
                               tf.keras.layers.Dense(units=256,activation='relu'),
                               BatchNormalization(),

                               tf.keras.layers.Dropout(0.5),
                               
                               #keras.layers.Dense(units=2,activation='softmax')
                               tf.keras.layers.Dense(units=2,activation='softmax')

])


model.compile(optimizer=Adam(lr=0.0001),loss='sparse_categorical_crossentropy',metrics=['accuracy'])

model_path = '/gdrive/MyDrive/Colab Notebooks/horse-or-human/horse_or_human_predictor2.h5'
checkpoint = ModelCheckpoint(model_path, monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')
callbacks_list = [checkpoint]
 
````

````
history=model.fit(training_data,
          epochs=100,
          verbose=1,
          validation_data=validation_data,
          callbacks=callbacks_list)
````

**Summary for Accuracy**

````
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model_accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')


````

````
def pred_human_horse(model, horse_or_human):
  test_image = image.load_img(horse_or_human, target_size = (150, 150))
  test_image = image.img_to_array(test_image)/255
  test_image = np.expand_dims(test_image, axis = 0)
 
  result = model.predict(test_image).round(3)
 
  pred = np.argmax(result)
  print(result, "--->>>", pred)
 
  if pred == 0:
    print('Predicted>>> Horse')
  else:
    print('Predicted>>> Human')
````

````
for horse_or_human in [h1,h2,h3,h4]:
  pred_human_horse(model1, horse_or_human)
````
